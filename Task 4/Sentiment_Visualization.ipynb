```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
from textblob import TextBlob
```

```python
train_data = pd.read_csv("C:\\Desktop\\docs\\Docs\\Task 4\\Dataset\\twitter_training.csv" , header = None , names = ['Id' , 'Game' , 'Sentiment' , 'Text'])
```

```python
validation_data = pd.read_csv("C:\\Desktop\\docs\\Docs\\Task 4\\Dataset\\twitter_validation.csv" , header = None , names = ['ID' , 'Game' , 'Sentiment' , 'Text'])
```

```python
print(train_data.head())
```

```python
print(validation_data.head())
```

```python
train_data['Sentiment_Polarity'] = train_data['Text'].apply(lambda x: TextBlob(str(x)).sentiment.polarity)
```

```python
plt.figure(figsize = (10 , 6))
sns.histplot(train_data['Sentiment_Polarity'] , bins  = 30 , kde = True)
```

```python
sns.boxplot(x = train_data['Sentiment_Polarity'])
```

```python
validation_data['Sentiment_Polarity'] = validation_data['Text'].apply(lambda x:TextBlob(str(x)).sentiment.polarity)
```

```python
sns.histplot(validation_data['Sentiment_Polarity'] , bins = 30 , kde = True)
```

```python
sns.boxplot(x = validation_data['Sentiment_Polarity'])
```

```python
sns.kdeplot(train_data['Sentiment_Polarity'])
sns.kdeplot(validation_data['Sentiment_Polarity'])
```

```python
plt.subplot(1,2,1)
sns.histplot(train_data['Sentiment_Polarity'] , bins = 15 , kde = True)
plt.subplot(1,2,2)
sns.histplot(validation_data['Sentiment_Polarity'] , bins = 15 , kde = True)
```
